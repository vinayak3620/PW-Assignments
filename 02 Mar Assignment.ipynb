#### Q2, Scatter  plot is used to just plot the datapoint with no line connecting the datapoints.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X_values")
plt.ylabel("Y_values")
plt.title("Scatter plot of x and y")
```

#### Q3. Subplot is used to make more than one chart in the same figure.

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.subplot(2,2,2)
plt.plot(x2,y2)

plt.subplot(2,2,3)
plt.plot(x3,y3)

plt.subplot(2,2,4)
plt.plot(x4,y4)
```

#### Q4. Bar plot is used in case the values are categorical values.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
```

```python
plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
```
